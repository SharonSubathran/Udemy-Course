```python
df = pandas.read_csv("supermarkets.csv")
```

```python
df
```

```python
#geocode expects st name, city and state with po box
```

```python
df["Address"]= df["Address"]+ "," +df["City"]+ "," +df["State"] + "," + df["Country"]
df
```

```python
#we did this to convert address to th form acceptable by gecoders
```

```python
#now to apply geocoders to this new column of addreses in the form we want-
```

```python
df["Coordinates"] = df["Address"].apply(nom.geocode)
```

```python
df
```

```python
df["Coordinates"][0]
```

```python
df["Coordinates"][0].latitude
```

```python
df["Latitude"] = df["Coordinates"].apply(lambda x: x.latitude if x != None else None)
```

```python
#above, lamda is used to apply the latitude function to each value of the coordinates column (like we did coordinates[0])
```
